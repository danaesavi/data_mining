#ID3 Algorithm


```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
base =pd.read_csv('car_categorical.csv',sep=',')
```

```python
base.head()
```

```python
target='car_eval'
attribute='buying'
```

```python
def entropy(base,target):
    entr=0
    total=len(base)
    for values in base[target].unique():
        num_value=len(base[base[target]==values])
        entr=entr-num_value/float(total)*np.log2(num_value/float(total))
    return entr
```

```python
def information_gain(base,target,attribute):
    total=len(base)
    inf_gain=entropy(base,target)
    for value in base[attribute].unique():
        sub_base=base[base[attribute]==value]
        inf_gain=inf_gain-float(len(sub_base))/total*entropy(sub_base,target)
    return inf_gain
```

```python
inf_gain=information_gain(base,target,attribute)
```

```python
attributes_list=['buying','maint','doors','persons','lug_boot','safety']
def tree(base,target,attributes_list,str_print):
    if len(base[target].unique())==1:   #all targets value are the same
        return base[target].unique()    #return the value
    elif len(attributes_list)== 0:      #we've analyzed every attribute
        return Counter(base[target]).most_common(1)[0][0]
    else:
        inf_gain=[]
        for attr in attributes_list:    #calculate the inf_gain for each attribute
            inf_gain.append(information_gain(base,target,attr))
        if max(inf_gain)!=0:            
            attr_ind=inf_gain.index(max(inf_gain)) # get the index of the max(inf_gain)
            
            modified_list=attributes_list[:]       #make a copy
            sel_attr=modified_list.pop(attr_ind)   #take out the max inf_gain attribute
            str_print=str_print +'\t'
            for value in base[sel_attr].unique():   
                print str_print+sel_attr+' '+value+' '+str(tree(base[base[sel_attr]==value],target,modified_list,str_print))
        else:
            return Counter(base[target]).most_common(1)[0][0]
    


    
```

```python
tree(base,'car_eval',attributes_list,' ')
```
